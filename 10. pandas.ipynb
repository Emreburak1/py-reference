### Where: loc[boolean list with same length as number of rows of data frame]. Return the row i in a dataframe if bool_list[i] == True. Row filtering.

```python
bool_list = [i%2 == 0 for i in range(len(df.index))]
df.loc[bool_list] 
```

```python
df.loc[[i % 2 == 0 for i in range(800)]] 
```

### df["Type 1"] returns a "series" of strings, which can be checked for the condition == "Fire" elementwise, which will return a list of bools. In the end, df.loc[bool_list]

```python
df.loc[df["Type 1"] == "Fire"] 
```

```python
df.loc[(df["Type 1"] == "Grass") & (df["Type 2"] == "Poison")]
```

```python
df.loc[((df["Type 1"] == "Grass") | (df["Type 2"] == "Poison")) & (df["HP"] > 70)]   
```

### Mutating - creating a new column with using the values in other columns. All of these methods below take column(s), do stuff with them, and return only them, as series. If they are not kept in original dataframe, then they are lost. But when they are kept with original dataframe, then original dataframe is now changed.

```python
df.loc[df["Type 1"] == "Fire", "Type 1"] = "FIRE"
df.loc[df["Type 1"] == "FIRE"].head()
```

```python
df.loc[df["Type 1"] == "FIRE", "Type 1"] = "Fire"
```

```python
df.loc[df["HP"] > 50, ["Super 1", "Super 2"]] = [True, False]
df.head()
```

```python
df.dtypes 
```

```python
df["Total"] = df.iloc[:, 4:10].sum(axis=1)  
df.head()
```

```python
df["Sqrt"] = sqrt(df[["Total"]])  
df.head()
```

```python
df["deneme"] = pd.Series(map(add, df["HP"], df["Attack"])) 
df   
```

### apply is more or less an equivalent to above cell. It operates row wise, so it is slower than vectorized functions.

```python
df.apply(lambda x: nv_add(x["HP"], x["Attack"]), axis=1) 
```

### assign takes the original dataframe, copies it, does stuff with copy, returns copied dataframe (not series, whole dataframe), original dataframe remains intact. It can be used with method chaining. new_column_name = (callable(df)) or a series. callable takes the whole dataframe as input and returns series, so it must be vectorized. When function is not vectorized by itself, it can be made one by map()

```python
# df.assign(deneme = lambda x: nv_add(x["HP"], x["Attack"]))
# ValueError: The truth value of a Series is ambiguous.
```

```python
df.assign(deneme2 = pd.Series(map(nv_add, df["HP"], df["Attack"]))) # not vectorized function, map is used to vectorize it, "= series" 
```

```python
df.assign(deneme = (lambda x: add(x["HP"], x["Attack"]))) # vectorized function, "= callable"  
```

```python
df.assign(deneme = df["HP"] + df["Attack"]).head()
```
